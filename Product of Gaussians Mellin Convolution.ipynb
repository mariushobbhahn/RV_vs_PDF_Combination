```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from scipy.integrate import quad
```

```python
m1, s1 = -3, 1
m2, s2 = 4, 2

x1 = np.linspace(-3, 3, 1000+1)
x2 = np.linspace(-5, 8, 1000+1)

x_prod = np.linspace(-8, 10, 1000+1)

num_samples = 10000
num_bins = int(np.sqrt(num_samples))

#plot one
norm1_pdf = norm.pdf(x_prod, m1, s1)
norm2_pdf = norm.pdf(x_prod, m2, s2)
norm1_samples = norm.rvs(m1, s1, size=num_samples)
norm2_samples = norm.rvs(m2, s2, size=num_samples)
norm1_samples_many = norm.rvs(m1, s1, size=num_samples*100)
norm2_samples_many = norm.rvs(m2, s2, size=num_samples*100)

#plot two
#norm_sum_pdf = norm.pdf(x_prod, m_sum, s_sum)
norm_prod_samples = norm1_samples * norm2_samples


#plot three: product of pdfs
norm_pdf_prod = np.array([norm.pdf(x_, m1, s1) * norm.pdf(x_, m2, s2) for x_ in x_prod])
```

```python
# plot the convolution of the two Gaussians

x_ = np.linspace(-30, 10, 1000+1)
norm2 = norm.pdf(x_, m2, s2)

total_frames = 100

int_list = []
m_list = []

fig, ax = plt.subplots(figsize=(10,5))
plt.close()

def animate_func(i):
    
    ax.cla()
    ax.plot(x_, norm2, color='orange')
    
    m_ = -30 + (10+30)*i/total_frames
    norm1 = norm.pdf(x_, m_, s1)
    ax.plot(x_, norm1, color='blue')
    f = lambda x: 1/x * norm.pdf(m_/x, m1, s1) * norm.pdf(x, m2, s2)
    integral, _ = quad(f, 0.00001, 100)
    int_list.append(integral)
    m_list.append(m_)
    for i in range(len(int_list)):    
        ax.axvline(m_list[i], 0.05, 0.05+int_list[i]/0.44, color='firebrick')
```

```python
from matplotlib import animation
from IPython.display import HTML

# Animation setup
anim = animation.FuncAnimation(
    fig, func=animate_func, frames=total_frames, interval=100, blit=False
)
HTML(anim.to_jshtml())
```
